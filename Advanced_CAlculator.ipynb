```python
import math

class Calculator:
    def __init__(self):
        # Dictionary to store operation symbol as key and function as value
        self.operations = {
            '+': lambda x, y: x + y,
            '-': lambda x, y: x - y,
            '*': lambda x, y: x * y,
            '/': self.divide  # Use a method for division to handle zero division
        }

    def add_operation(self, symbol, function):
        """Add a new operation and its function to the dictionary."""
        self.operations[symbol] = function

    def divide(self, x, y):
        if y == 0:
            raise ValueError("Cannot divide by zero!")
        return x / y

    def calculate(self, num1, operator, num2=None):
        # Error handling for input types
        if not (isinstance(num1, (int, float)) and (num2 is None or isinstance(num2, (int, float)))):
            print("Error: Please enter valid numbers.")
            raise TypeError("Invalid number input.")

        if operator not in self.operations:
            print(f"Error: Operation '{operator}' is not supported.")
            raise ValueError("Invalid operation.")

        # For unary operations like square root and logarithm
        if num2 is None:
            return self.operations[operator](num1)
        else:
            return self.operations[operator](num1, num2)

# Advanced operation functions
def power(x, y):
    return x ** y

def square_root(x):
    if x < 0:
        raise ValueError("Cannot take the square root of a negative number.")
    return math.sqrt(x)

def logarithm(x):
    if x <= 0:
        raise ValueError("Cannot take the logarithm of zero or negative number.")
    return math.log(x)

# Main program
calc = Calculator()
# Add advanced operations
calc.add_operation('^', power)
calc.add_operation('sqrt', square_root)
calc.add_operation('log', logarithm)

print("Welcome to the Advanced Calculator!")
print("Available operations: +, -, *, /, ^ (power), sqrt (square root), log (natural logarithm)")
print("Type 'exit' to quit.\n")

while True:
    op = input("Enter operation symbol (+, -, *, /, ^, sqrt, log) or 'exit': ")
    if op == 'exit':
        print("Goodbye!")
        break

    try:
        if op in ['sqrt', 'log']:
            n1 = float(input("Enter the number: "))
            result = calc.calculate(n1, op)
        else:
            n1 = float(input("Enter the first number: "))
            n2 = float(input("Enter the second number: "))
            result = calc.calculate(n1, op, n2)
        print("Result:", result)
    except Exception as e:
        print("Calculation error:", e)
```
