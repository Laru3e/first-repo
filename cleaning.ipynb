```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import MinMaxScaler
```

```python
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns= diabetes.feature_names)
df['target']= diabetes.target
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
threshold = df['target'].quantile(0.99)
outliers = df[df['target'] > threshold]
outliers
```

```python
outliers.shape[0]
```

```python
df.head()
```

```python
from typing import ValuesView
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop('target', axis=1))
df_scaled = pd.DataFrame(scaled_features, columns=diabetes.feature_names)
df_scaled['target']=df['target'].values
df_scaled
```

```python
#standardscaler transform features to have a mean of 0 and standard deviation of 1
```

```python
minmax = MinMaxScaler()
scaled_features_mm = minmax.fit_transform(df.drop('target', axis=1))
df_scaled_mm = pd.DataFrame(scaled_features_mm, columns= diabetes.feature_names)
df_scaled_mm['target']= df['target'].values
df_scaled_mm
```
